```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

boroughs = ["Queens","Manhattan","Bronx","Brooklyn","Staten Island","New York City"]
df = pd.read_csv("MedianIncomes.csv", on_bad_lines='skip')
df = df[~df["Location"].isin(boroughs)]
df = df[df["Household Type"].isin(["All Households"])]
df = df[df.TimeFrame == 2019].filter(items=["Location","Data"])
df.head()
```
